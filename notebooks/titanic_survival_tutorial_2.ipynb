```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
men = train_data.loc[train_data.Sex == 'male']['Survived']
rate_men = sum(men) / len(men)
rate_men
```

```python
def plot_pie(title, rate, color='dark'):
    keys = ['Survived', 'Died']
    data = [rate, 1-rate]
    
    palette_color = sns.color_palette(color)
    
    plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
    
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_pie('Survival Rate Men', rate_men)
```

```python
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

first = train_data.loc[train_data.Pclass == 1]["Survived"]
rate_1st = sum(first)/len(first)

second = train_data.loc[train_data.Pclass == 2]["Survived"]
rate_2nd = sum(second)/len(second)

third = train_data.loc[train_data.Pclass == 3]["Survived"]
rate_3rd = sum(third)/len(third)
```

```python
plot_pie('% of Women who Survived', rate_women, color='Paired_r')
plot_pie('% of 1st class passengers who Survived', rate_1st, color='Pastel1')
plot_pie('% of 2nd class passengers who Survived', rate_2nd, color='CMRmap')
plot_pie('% of 3rd class passengers who Survived', rate_3rd, color='BuPu')
```

```python
modeling_data = train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
modeling_data.info()
```

```python
from sklearn.impute import SimpleImputer
mode_imputer = SimpleImputer(strategy='most_frequent')
```

```python
modeling_data['Age'] = mode_imputer.fit_transform(modeling_data['Age'].values.reshape(-1,1))
```

```python
modeling_data.info()
```

```python
modeling_data['Embarked'] = modeling_data['Embarked'].fillna(method='bfill')
```

```python
modeling_data.info()
```

```python
modeling_data
```

```python
encoded_data = pd.get_dummies(modeling_data, columns=['Embarked', 'Sex'])
encoded_data
```

```python
X = encoded_data.drop('Survived', axis=1)
Y = encoded_data['Survived']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

knn = KNeighborsClassifier(n_neighbors=7)

X_train, X_test, Y_train, Y_test = train_test_split(
             X, Y, test_size = 0.2, random_state=37)

knn.fit(X_train, Y_train)

print(knn.score(X_test, Y_test))
```

```python
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=7)
scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy')

print(scores)

print(scores.mean())
```

```python
neighbors = [i for i in range(1,30)]

cv_scores = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy')
    cv_scores.append(scores.mean())
    
accuracy = [x for x in cv_scores]
optimal_k_index = accuracy.index(max(accuracy))
optimal_k = neighbors[optimal_k_index]
print(optimal_k)

# plot misclassification error vs k
plt.plot(neighbors, accuracy)
plt.xlabel('Number of Neighbors K')
plt.ylabel('Accuracy')
plt.show()
```
