```python
def plot_pie(title, rate, color='dark'):
    keys = ['Suvived', 'Died']
    data = [rate, 1-rate]
    
    palette_color = sns.color_palette(color)
    
    plt.pie(data, labels=keys, colors=palette_color, autopct='%.2f%%')
    
    plt.title(title)
    plt.legend()
    
    plt.show()
```

```python
plot_pie('Survival rate of men', rate_men)
```

```python
encoded_data = pd.get_dummies(modeling_df, columns=['Embarked', 'Sex'])
encoded_data
```

```python
X = encoded_data.drop('Survived', axis=1)
Y = encoded_data['Survived']
```

```python
Y
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
                X, Y, test_size=0.2, random_state=25)

knn = KNeighborsClassifier(n_neighbors=6)

knn.fit(X_train, Y_train)

print(knn.score(X_test, Y_test))
```

```python
from sklearn.metrics import confusion_matrix
preds = knn.predict(X_test)

cm = confusion_matrix(Y_test, preds)
tn, fp, fn, tp = cm.ravel()

```

```python
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=6)

scores = cross_val_score(knn, X, Y, cv=10, scoring='accuracy')

print(scores)
print(scores.mean())
```

```python
n_neighbors = [i for i in range(1,30)]

accuracy = []
for k in n_neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, scoring='accuracy')
    accuracy.append(scores.mean())
    
accuracy
```

```python
max_accuracy = max(accuracy)
optimal_k_index = accuracy.index(max_accuracy)
optimal_k = n_neighbors[optimal_k_index]
print(optimal_k)

plt.plot(n_neighbors, accuracy)
plt.xlabel('Number of neighbors K')
plt.ylabel('Accuracy')
plt.show()
```
