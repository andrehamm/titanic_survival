```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
men = train_data.loc[train_data.Sex == 'male']['Survived']
rate_men = sum(men)/len(men)
rate_men
```

```python
women = train_data.loc[train_data.Sex == 'female']['Survived']
rate_women = sum(women)/len(women)

first = train_data.loc[train_data.Pclass == 1]["Survived"]
first_rate = sum(first)/len(first)

second = train_data.loc[train_data.Pclass == 2]["Survived"]
second_rate = sum(second)/len(second)

third= train_data.loc[train_data.Pclass == 3]["Survived"]
third_rate = sum(third)/len(third)
```

```python
def plot_pie(title, rate, color='dark'):
    keys = ['Survived', 'Died']
    data = [rate, 1-rate]
    palette_color = sns.color_palette(color)
    plt.pie(data, labels=keys, colors=palette_color, autopct='%.0f%%')
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
plot_pie('% of men who Survived', rate_men)
```

```python
plot_pie('% of Women who Survived', rate_women)
plot_pie('Survival rate of 1st class passengers', first_rate, color='pastel')
plot_pie('Survival rate of 2nd class passengers', second_rate, color='CMRmap')
plot_pie('Survival rate of 3rd class passengers', third_rate, color='BuPu')
```

```python
first_class_women = train_data.loc[train_data.Sex == 'female'][train_data.Pclass == 1]['Survived']
rate_1st_women = sum(first_class_women)/len(first_class_women)
plot_pie('Survival rate of 1st class women', rate_1st_women)
```

```python
modeling_data = train_data.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
modeling_data.info()
```

```python
from sklearn.impute import SimpleImputer
mode_imputer = SimpleImputer(strategy='most_frequent')
```

```python
modeling_data['Age'] = mode_imputer.fit_transform(modeling_data['Age'].values.reshape(-1,1))
```

```python
modeling_data['Embarked'] = modeling_data['Embarked'].fillna(method='bfill')
```

```python
modeling_data
```

```python
encoded_data = pd.get_dummies(modeling_data, columns=['Sex', 'Embarked'])
encoded_data
```

```python
X = encoded_data.drop('Survived', axis=1)
Y = encoded_data['Survived']
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.2, random_state=25)

knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(X_train, Y_train)

print(knn.score(X_test, Y_test))
```

```python
from sklearn.model_selection import cross_val_score

knn = KNeighborsClassifier(n_neighbors=7)

scores = cross_val_score(knn, X, Y, cv = 10, scoring='accuracy')

print(scores)
print(scores.mean())
```

```python
neighbors = [i for i in range(1,30)]

cv_scores = []
for k in neighbors: 
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X, Y, cv = 10, scoring='accuracy')
    cv_scores.append(scores.mean())
    
print(cv_scores[:5])
```

```python
max_score_index = cv_scores.index(max(cv_scores))
max_score_neighbor = neighbors[max_score_index]

print(max_score_neighbor)

plt.plot(neighbors, cv_scores)
plt.xlabel('NUumber of Neghbors K')
plt.ylabel('Accuracy')
plt.show()
```
